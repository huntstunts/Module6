```python
# Import the required libraries and dependencies
import pandas as pd
import hvplot.pandas
from pathlib import Path
```

```python
 # Using the read_csv function and Path module, create a DataFrame 
# by importing the sfo_neighborhoods_census_data.csv file from the Resources folder
sfo_data_df = pd.read_csv(
    Path('sfo_neighborhoods_census_data.csv')
)

# Review the first and last five rows of the DataFrame
sfo_data_df.head()
```

```python
#Made different cell for last five rows
#Having first and last five rows in same cell only produced last five rows
sfo_data_df.tail()
```


Step 1: 
Use the groupby function to group the data by year. Aggregate the 
 results by the mean of the groups.

```python
sfo_data_df.dtypes
```

```python

# Create a numerical aggregation that groups the data by the year and then averages the results.

housing_units_by_year=sfo_data_df.drop(columns=['sale_price_sqr_foot','gross_rent'])
housing_units_by_year = housing_units_by_year.groupby('year').mean('housing_units')

# Review the DataFrame
housing_units_by_year.head(10)
```

 
 Step 2: 
 Use the hvplot function to plot the housing_units_by_year DataFrame as a bar chart. Make the x-axis represent the year and the y-axis represent the housing_units.


Step 3: 
Style and format the line plot to ensure a professionally styled visualization.

```python
housing_units_by_year.hvplot.bar(
    rot=90
).opts(
    yformatter='%.0f'
)
```

```python
# Create a visual aggregation explore the housing units by year
housing_units_by_year.hvplot.line(
    label="Number of Housing Units 2010 through 2016",
    rot=45,
).opts(
    yformatter='%.0f',
    line_color='orange',
    hover_line_color='blue'
)
    
```
